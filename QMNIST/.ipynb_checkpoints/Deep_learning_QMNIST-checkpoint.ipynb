```python
import pickle
import matplotlib.pyplot as plt
```

```python
# Load files into memory
def unpickle(file):
    """
    Loads pickle file and returns it.
    """
    with open(file, 'rb') as fo:
        file = pickle.load(fo, encoding='bytes')
    return file

qmnist = unpickle("Data/MNIST-120k")
```

```python
data = qmnist['data']
labels = qmnist['labels']
```

```python
for i in range(0, 10):
    plt.imshow(data[i])
    plt.title(str(labels[i]))
    plt.show()
```
